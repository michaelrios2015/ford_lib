```python
# going to just keep a few title and

def format_data(filename):

    formatted = []

    # Opening our text file in read only 
    # mode using the open() function 
    # with open('LC.txt', 'r') as file: 

    file = open(filename, 'r', encoding="utf8")

    # print(file)

    data = file.readlines() 

    # print(data)

      # this keeps track of where i am in the new list 
    counter = 0

    for ind, line in enumerate(data):

    #         print(ind)
    #         print(line)
        if "Title:" in line: 
          # print(line)
          # print("--------------")
          line = line.strip()
          if line[-1] == ",":
              line = line[:-1]

          formatted.append(line)
          if ind > 0:
            # print(counter)
            # print((data[ind-2]))
            formatted[counter] += " |" + data[ind-2].strip()
            counter += 1   


    formatted[counter] += " |" + data[-1].strip()

#     print(formatted[0])
    
    return formatted


# with open(r'LCformatted.txt', 'w') as fp:
#   for item in formatted:
#       # write each item on a new line
#       fp.write("%s\n" % item)
#   print('Done')

# import pandas as pd

# df = pd.read_csv('RHformatted.txt', sep='|')

# print(df[0])
```

```python
lc = format_data("LC.txt")
rh = format_data("RH.txt")
lc[0]
```

```python
import pandas as pd

def to_df(df):

    new_df = pd.DataFrame(df)

#     new_df.head()

    new_df[['title', 'cn']] = new_df[0].str.split('|', expand=True)

    new_df = new_df.drop([0], axis=1)

    return new_df

lc_df = to_df(lc)
rh_df = to_df(rh)

rh_df.head(6)
```
